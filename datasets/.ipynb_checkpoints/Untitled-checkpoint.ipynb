```python
class VOC2012Segmentation(Dataset):
    
    class_names = np.array([
        'background',
        'aeroplane',
        'bicycle',
        'bird',
        'boat',
        'bottle',
        'bus',
        'car',
        'cat',
        'chair',
        'cow',
        'diningtable',
        'dog',
        'horse',
        'motorbike',
        'person',
        'potted plant',
        'sheep',
        'sofa',
        'train',
        'tv/monitor',
    ])
    mean_bgr = np.array([104.00698793, 116.66876762, 122.67891434])

    def __init__(self, dataset_dir, split = 'train', transform = False):
        self.dataset_dir = dataset_dir
        self._transform = transform
        
        if self._transform:
            self.transform = transforms.Compose([
                transforms.Resize((480, 520)),  # Resize to 256x256
                transforms.ToTensor(),           # Convert PIL image to PyTorch tensor (C x H x W)
                transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])  # Normalize the image
            ])
        else:
            self.transform = transforms.Compose([
                transforms.Resize((480, 520)),  # Resize to 256x256
                transforms.ToTensor(),           # Convert PIL image to PyTorch tensor (C x H x W)
            ])
        
        self.split = split

        imgset_files = os.path.join(self.dataset_dir, 
                                      f'ImageSets/Segmentation/{self.split}.txt')

        self.img_files = []
        self.segmentation_files = []
        with open(imgset_files, 'r') as f:              
                
            for label in f:
                label = label.strip()
                img_files = os.path.join(self.dataset_dir, f'JPEGImages/{label}.jpg')
                segmentation_files = os.path.join(self.dataset_dir, f'SegmentationClass/{label}.png')
                self.img_files.append(img_files)
                self.segmentation_files.append(segmentation_files)
    
    def __getitem__(self, index):
        img_file = self.img_files[index]
        segmentation_files = self.segmentation_files[index]

        # Load the image
        img = Image.open(img_file)
        img = self.transform(img)

        # Load the label
        segment_img = Image.open(segmentation_files)
        segment_img = np.array(segment_img, dtype=np.int32)
        # resize_transform = transforms.Compose([
        #     transforms.Resize((480, 520)),
        #     transforms.ToTensor(),
        # ])
        # segment_img = resize_transform(segment_img)
        segment_img[segment_img == 255] = -1
        segment_img = torch.from_numpy(segment_img)
        # segment_img = segment_img.squeeze(0)
        
        return img, segment_img

    def __len__(self):
        return len(self.img_files)
        
            
        
```

```python
class SBDClassSeg(VOC2012Segmentation):
   

    def __init__(self, dataset_dir, split='train', transform=False):
        self.dataset_dir = dataset_dir
        self.split = split
        self._transform = transform

        if self._transform:
            self.transform = transforms.Compose([
                transforms.Resize((480, 520)),  # Resize to 256x256
                transforms.ToTensor(),           # Convert PIL image to PyTorch tensor (C x H x W)
                transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])  # Normalize the image
            ])
        else:
            self.transform = transforms.Compose([
                transforms.Resize((480, 520)),  # Resize to 256x256
                transforms.ToTensor(),           # Convert PIL image to PyTorch tensor (C x H x W)
            ])
        
        imgset_files = os.path.join(self.dataset_dir, 
                                      f'{self.split}.txt')
        # for split in ['train', 'val']:
        #     imgsets_file = os.path.join(dataset_dir, '%s.txt' % split)
        #     for did in open(imgsets_file):
        #         did = did.strip()
        #         img_file = os.path.join(dataset_dir, 'img/%s.jpg' % did)
        #         lbl_file = os.path.join(dataset_dir, 'cls/%s.mat' % did)
        #         self.files[split].append({
        #             'img': img_file,
        #             'lbl': lbl_file,
        #         })

        self.img_files = []
        self.segmentation_files = []
        with open(imgset_files, 'r') as f:              
            for label in f:
                label = label.strip()
                img_files = os.path.join(self.dataset_dir, f'img/{label}.jpg')
                segmentation_files = os.path.join(self.dataset_dir, f'cls/{label}.,at')
                self.img_files.append(img_files)
                self.segmentation_files.append(segmentation_files)
                
    def __getitem__(self, index):
        img_file = self.img_files[index]
        segmentation_files = self.segmentation_files[index]

        # load image
        img = Image.open(img_file)
        img = np.array(img, dtype=np.uint8)
        
        # load label
        lbl_file = segmentation_files
        mat = io.loadmat(lbl_file)
        lbl = mat['GTcls'][0]['Segmentation'][0].astype(np.int32)
        lbl[lbl == 255] = -1
        
        # if self._transform:
        #     return self.transform(img, lbl)
        # else:
        return img, lbl
```
